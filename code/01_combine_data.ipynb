````python
import pandas as pd
import glob
import os
import re

# Folder containing all CSV files (update path if needed)
folder_path = "."  # current directory; change to your folder path
file_pattern = "Raw_data_1Day_*.csv"
file_list = glob.glob(os.path.join(folder_path, file_pattern))

# Dictionary to rename columns (remove units and special characters)
column_rename = {
    "Timestamp": "timestamp",
    "PM2.5 (Âµg/mÂ³)": "pm2_5",
    "PM10 (Âµg/mÂ³)": "pm10",
    "NO (Âµg/mÂ³)": "no",
    "NO2 (Âµg/mÂ³)": "no2",
    "NOx (ppb)": "nox",
    "NH3 (Âµg/mÂ³)": "nh3",
    "SO2 (Âµg/mÂ³)": "so2",
    "CO (mg/mÂ³)": "co",
    "Ozone (Âµg/mÂ³)": "ozone",
    "Benzene (Âµg/mÂ³)": "benzene",
    "Toluene (Âµg/mÂ³)": "toluene",
    "Xylene (Âµg/mÂ³)": "xylene",
    "O Xylene (Âµg/mÂ³)": "o_xylene",
    "Eth-Benzene (Âµg/mÂ³)": "ethylbenzene",
    "MP-Xylene (Âµg/mÂ³)": "mp_xylene",
    "AT (°C)": "at",
    "RH (%)": "rh",
    "WS (m/s)": "ws",
    "WD (deg)": "wd",
    "RF (mm)": "rf",
    "TOT-RF (mm)": "tot_rf",
    "SR (W/mt2)": "sr",
    "BP (mmHg)": "bp",
    "VWS (m/s)": "vws"
}

def extract_station_name(filename):
    """Extract station name from filename like 
    Raw_data_1Day_2021_site_125_Punjabi_Bagh_Delhi_DPCC_1Day.csv
    """
    basename = os.path.basename(filename)
    # Pattern: ..._site_<number>_<StationName>_Delhi_DPCC_...
    # Station name may contain underscores
    match = re.search(r"_site_\d+_(.+?)_Delhi_DPCC", basename)
    if match:
        return match.group(1)
    else:
        # fallback: split by underscores and take part after site number
        parts = basename.split('_')
        # find index of 'site'
        try:
            site_idx = parts.index('site')
            # station name is the next part after the site number
            station = parts[site_idx + 2]
            # if there are more parts before Delhi, join them? Actually station name can have underscores
            # Better: after site number, collect until 'Delhi'
            idx = site_idx + 2
            station_parts = []
            while idx < len(parts) and parts[idx] != 'Delhi':
                station_parts.append(parts[idx])
                idx += 1
            return '_'.join(station_parts)
        except (ValueError, IndexError):
            return "unknown"

# List to hold dataframes
dfs = []

for file in file_list:
    print(f"Processing {file}")
    try:
        df = pd.read_csv(file)
        
        # Rename columns
        df.rename(columns=column_rename, inplace=True)
        
        # Add station name
        station = extract_station_name(file)
        df['station'] = station
        
        dfs.append(df)
    except Exception as e:
        print(f"Error reading {file}: {e}")

# Combine all
if dfs:
    combined = pd.concat(dfs, ignore_index=True)
    
    # Convert timestamp to datetime if possible
    if 'timestamp' in combined.columns:
        combined['timestamp'] = pd.to_datetime(combined['timestamp'], errors='coerce')
    
    # Save to CSV
    output_file = "combined_delhi_pollution_2021_2025.csv"
    combined.to_csv(output_file, index=False)
    print(f"Saved combined data to {output_file}")
    print(f"Total rows: {len(combined)}")
else:
    print("No CSV files found.")
````

# Data Quality Investigation: Missing 2023–2024 Data in Daily Aggregates

## 1. Observation (EDA Phase)

During exploratory data analysis, we noticed that annual average PM2.5 tables showed only years **2021, 2022, and 2025** for all stations. Years 2023 and 2024 were completely absent.

```text
Annual average PM2.5 (µg/m³) per station:
year              2021   2022   2025
station                             
Jahangirpuri     135.8  127.1  130.8
Wazirpur         119.9  114.3  124.3
...
```

The daily combined file (`combined_1dayFreq_delhi_pollution_2021_2025.csv`) had only 9,855 rows instead of the expected ~16,425 (9 stations × 5 years × 365 days).

## 2. Initial Hypothesis

We suspected a timestamp parsing issue because the raw daily CSV files for 2023–2024 might use a different date format (`DD-MM-YYYY`) compared to 2021–2022 (`YYYY-MM-DD`). The combination script used a fixed format or default parser without `dayfirst`, causing many rows to become `NaT` and then dropped.

## 3. Investigation

### 3.1 Checking raw daily files

Opening a raw daily file (e.g., `Raw_data_1Day_2023_site_125_Punjabi_Bagh_Delhi_DPCC_1Day.csv`) in Notepad++ showed:

```
2023-07-13,63.51785714,123.6153846,...
2023-07-14,,,,,,,,,,0.0,,,
2023-07-15,,,,,,,,,,0.0,,,
...
```

- The timestamp format is **`YYYY-MM-DD`** – same as 2021–2022.
- Many rows have only a timestamp and zeros (missing pollutant data), but the timestamp itself is valid.

### 3.2 Running the combination script with verbose parsing

The combination script originally contained this logic:

```python
if year >= 2023:
    df['timestamp'] = pd.to_datetime(df['timestamp'], dayfirst=True, errors='coerce')
else:
    df['timestamp'] = pd.to_datetime(df['timestamp'], errors='coerce')
```

For 2023–2024 files, `dayfirst=True` was applied. This caused strings like `2023-07-14` to be interpreted as **day=2023, month=7, year=14** → invalid → `NaT`. Consequently, **all rows** in those files became invalid and were dropped.

### 3.3 Counting dropped rows

After combination, the script reported:

```
Processing Raw_data_1Day_2023_site_125_Punjabi_Bagh_Delhi_DPCC_1Day.csv...
  Warning: 221 rows with invalid timestamp (will be dropped)
```

But in reality, **all 365 rows** were invalid; the “221” was an artifact of how the warning was counted. The final combined file had no rows from 2023–2024.

## 4. Root Cause

**Incorrect assumption about date format in raw daily files.**  
The files use `YYYY-MM-DD` consistently across all years. Forcing `dayfirst=True` for 2023–2024 broke parsing. The combination script should have used the **same default parser** for all files.

## 5. Solution

### 5.1 Fix the combination script

Remove the year‑based branching. Use `pd.to_datetime()` with no `dayfirst` argument (default parser works for `YYYY-MM-DD`).

**Corrected parsing function:**

```python
def parse_timestamp_column(df, col_name='timestamp'):
    if col_name in df.columns:
        df[col_name] = pd.to_datetime(df[col_name], errors='coerce')
    return df
```

### 5.2 Re‑run combination

After applying the fix, the combined CSV contained all 5 years with 16,434 rows (accounting for leap years). The timestamp column had zero `NaT` values.

### 5.3 Reprocess daily data

The corrected combined CSV was then passed through the daily cleaning pipeline (imputation, AQI calculation, etc.). The resulting parquet file now includes 2023 and 2024.

## 6. Impact

- **Annual PM2.5 averages** now show all five years (2021–2025).
- **Time series plots** for 2023–2024 are correctly displayed.
- **Outlier detection** and **forecasting models** now use the full 5‑year dataset, improving reliability.

## 7. Preventive Measures for Future

- Always **inspect a sample of raw files** before writing parsing logic – do not assume format changes.
- Use **consistent parsing** across all files unless you have verified different formats.
- Add **logging of year‑wise row counts** after combination to catch missing years early.
